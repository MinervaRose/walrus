## Exercise 1: Figures and Axes

Create a figure with 3 subplots in a row. In each, plot a histogram of `x`, `y`, and `z` diamond size in mm. Specify the x-title of each includes the units. For this exercise, use the `pyplot` interface, and the `plt.hist` function.

```python
# YOUR CODE HERE

plt.figure(figsize=(18, 5))

# --- Subplot 1: x dimension ---
plt.subplot(1, 3, 1)
plt.hist(df['x'], bins=30, edgecolor='black')
plt.xlabel("x size (mm)")
plt.ylabel("Count")
plt.title("Histogram of x dimension")

# --- Subplot 2: y dimension ---
plt.subplot(1, 3, 2)
plt.hist(df['y'], bins=30, edgecolor='black')
plt.xlabel("y size (mm)")
plt.title("Histogram of y dimension")

# --- Subplot 3: z dimension ---
plt.subplot(1, 3, 3)
plt.hist(df['z'], bins=30, edgecolor='black')
plt.xlabel("z size (mm)")
plt.title("Histogram of z dimension")

plt.tight_layout()
plt.show()
```

## Exercise 2: Figures and Axes

Create a figure with 2 subplots in a row. In the left subplot, plot the distribution of diamond quality; in the right, plot the distribution of clarity. Make sure to appropriately title the x- and y- axis of each plot. For this exercise, use the `axes` interface.

```python
# YOUR CODE HERE
import matplotlib.pyplot as plt

# Create a figure with 2 subplots in a row (Axes interface)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Left subplot: distribution of cut (diamond quality) ---
axes[0].hist(df['cut'])
axes[0].set_xlabel("Diamond Cut Quality")
axes[0].set_ylabel("Count")
axes[0].set_title("Distribution of Diamond Cut")

# --- Right subplot: distribution of clarity ---
axes[1].hist(df['clarity'])
axes[1].set_xlabel("Diamond Clarity")
axes[1].set_ylabel("Count")
axes[1].set_title("Distribution of Diamond Clarity")

plt.tight_layout()
plt.show()
```

## Exercise 3: Scales and Transformations

Create a histogram of diamond price using 1000 bins. You will notice that when using such a fine bin size, several bins are zero! This is likely due to marketing and consumer psychology. Identify which bins are zero by adjusting the x-axis limits.

```python
# YOUR CODE HERE
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 5))

# 1000 bins over the full price range
plt.hist(df['price'], bins=1000, edgecolor='none')

plt.xlabel("Price (USD)")
plt.ylabel("Count")
plt.title("Diamond Price Distribution (1000 bins)")

# Zoom in to identify zero-height bins (e.g., around typical psychological prices)
#  adjust these limits interactively
plt.xlim(0, 6000)   # adjust these numbers based on what you observe
# plt.xlim(2000, 4000)  # try narrower ranges to reveal individual zero bins

plt.show()
```

## Exercise 4: Scales and Transformations

Looking at a histogram of price, one notices a long skewed tail. Plot this using a `log` transformation.

```python
# YOUR CODE HERE
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.hist(df['price'], bins=100, edgecolor='black')
plt.xscale('log')     #  log transformation on the x-axis

plt.xlabel("Price (USD, log scale)")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices (Log-Transformed X-Axis)")

plt.tight_layout()
plt.show()
```
